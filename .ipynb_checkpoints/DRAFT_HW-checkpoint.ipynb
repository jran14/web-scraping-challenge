```python
#click each of the links to the hemispheres in order to find the image url to the full resolution image
astro_url='https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser = Browser('chrome', headless=True)
browser.visit(astro_url)
astro = bs(browser.html, 'html.parser')

hemispheres = browser.find_link_by_partial_text('Hemisphere')
for hemisphere in hemispheres:
    hemi_search= hemisphere.click()
    for hemi in hemi_search:
        #hemisphere name
        hemi_find=astro.select_one('div.container div.content')
        hemi_name =hemi_find.find("h2",class_="title").get_text()
        #hemisphere image url
        img_url=astro.select_one('div.container div.wide-image-wrapper img.wide-image')['src']
        
        print(hemi_name)
        print(img_url)
```
